### *As a first step I aim to provide the code for scraping the web page linked* [books.toscrape.com](https://books.toscrape.com/catalogue/category/books/humor_30/index.html)


```python
pip install requests beautifulsoup4

import requests
from bs4 import BeautifulSoup


url = 'https://books.toscrape.com/catalogue/category/books/humor_30/index.html'  
response = requests.get(url)
html_content = response.content
print(html_content)

 if response.status_code == 200:
    page_content = response.text
 else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
    
soup = BeautifulSoup(page_content, 'html.parser')


headings = soup.find_all('h3')
for heading in headings:
    print(heading.text)
```

### *And lastly I parse the contents as parts of speech*

```python
!pip install requests beautifulsoup4 nltk

import requests

from bs4 import BeautifulSoup
import nltk


nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
url = 'https://books.toscrape.com/catalogue/category/books/humor_30/index.html'


response = requests.get(url)


if response.status_code == 200:
    page_content = response.text
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")


soup = BeautifulSoup(page_content, 'html.parser')


book_titles = [title.get_text() for title in soup.find_all('h3')]
text_content = ' '.join(book_titles)
print(text_content)



tokens = nltk.word_tokenize(text_content)

pos_tags = nltk.pos_tag(tokens)



for word, pos in pos_tags:
    print(f'{word}: {pos}')
```
